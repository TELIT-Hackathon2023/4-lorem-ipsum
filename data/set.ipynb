```python
import json


with open("data.json", 'r') as file:
    js = json.load(file)


print(js["mediawiki"]["page"][0])

print(len(js["mediawiki"]["page"]))
```

```python

res = []
for i in js["mediawiki"]["page"]:
    if i["title"] not in res:
        res.append(i["title"])
print(res)
```

```python
no_user = []
exceptions = ["File:", "User:", "User blog", "User talk:", "Category:", "Template talk:", "MediaWiki:", "Talk:", "Template:", ':']
for i in res:
    flag = 1
    for j in exceptions:
        if j in i:
            flag = 0
            continue
    if flag == 1:
        no_user.append(i)

print(no_user)
print(len(no_user))
```

```python
import json

f = json.dumps({"titles":no_user}, indent=4)
with open("titles.json", 'w', encoding="utf-8") as fi:
    fi.write(f)
```
